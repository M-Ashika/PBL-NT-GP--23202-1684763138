```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Price'] = le.fit_transform(df['Price'])
```

```python
df.info()
```

```python
x = df.drop('Built Year',axis=1)
x.head()
```

```python
y = df['Built Year']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,test,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
reg = Sequential()
reg.add(Dense(4,activation='relu'))
reg.add(Dense(12,activation='relu'))
reg.add(Dense(32,activation='relu'))
reg.add(Dense(1,activation='linear'))
```

```python
reg.compile(optimizer='adam' ,loss='mse' ,metrics=['mse'])
```

```python
reg.fit(xtrain,ytrain,batch_size=10,epochs=25,validation_data=(test,ytest))
```
